```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("NFL Play by Play 2009-2016 (v3).csv", low_memory = False)
```

```python
for col in df.columns:
    print(col)
```

```python
relevant_columns = ['Date', 'GameID', 'Drive', 'qtr', 'down', 'time', 'yrdline100', 'ydstogo',
                   'GoalToGo', 'PosTeamScore', 'DefTeamScore', 'ScoreDiff', 'HomeTeam',
                   'AwayTeam', 'Timeout_Indicator', 'Timeout_Team', 'posteam_timeouts_pre',
                   'HomeTimeouts_Remaining_Pre', 'AwayTimeouts_Remaining_Pre',
                   'HomeTimeouts_Remaining_Post', 'AwayTimeouts_Remaining_Post',
                   'PlayAttempted', 'desc', 'PlayType', 'PassAttempt', 'RushAttempt',
                   'FieldGoalResult', 'FieldGoalDistance', 'Fumble', 'ReturnResult',
                   'Challenge.Replay', 'ChalReplayResult', 'Accepted.Penalty',
                   'PenaltyType', 'PenalizedTeam', 'PenalizedPlayer', 'Penalty.Yards',
                   'Passer', 'Passer_ID', 'Rusher', 'Rusher_ID', 'Receiver', 'Receiver_ID',
                   'Interceptor', 'Returner', 'BlockingPlayer', 'Tackler1', 'Tackler2',
                   'Yards.Gained', 'Touchdown', 'ExPointResult', 'TwoPointConv', 'Safety',
                   'Onsidekick', 'PuntResult', 'FieldGoalResult', 'No_Score_Prob',
                   'Opp_Field_Goal_Prob', 'Opp_Safety_Prob', 'Opp_Touchdown_Prob',
                   'Field_Goal_Prob', 'Safety_Prob', 'Touchdown_Prob', 'ExpPts', 'EPA',
                   'Home_WP_pre', 'Away_WP_pre', 'Home_WP_post', 'Away_WP_post', 'Win_Prob',
                   'WPA', 'airWPA', 'yacWPA', 'DefensiveTeam']

df = pd.read_csv("NFL Play by Play 2009-2016 (v3).csv", usecols=relevant_columns, low_memory = False)
```

```python
y = df['Touchdown']

x = df[['down', 'ydstogo', 'yrdline100', 'GoalToGo', 'PlayType', 'PassAttempt', 
       'RushAttempt', 'FieldGoalResult', 'ExpPts', 'EPA', 'Home_WP_pre', 'Away_WP_pre']]

x = pd.get_dummies(x, columns = ['PlayType', 'FieldGoalResult'])

x.fillna(0, inplace = True)

scaler = MinMaxScaler()

x[['ydstogo', 'yrdline100', 'ExpPts', 'EPA', 'Home_WP_pre', 'Away_WP_pre']] = scaler.fit_transform(x[['ydstogo', 'yrdline100', 'ExpPts', 'EPA', 'Home_WP_pre', 'Away_WP_pre']])

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 42)
```

```python
model = RandomForestClassifier(n_estimators = 100, random_state = 42)

model.fit(x_train, y_train)

y_pred = model.predict(x_test)

y_prob = model.predict_proba(x_test)

roc_auc = roc_auc_score(y_test, y_pred, multi_class = 'ovr')
print(f'ROC AUC Score: {roc_auc:.2f}')

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

report = classification_report(y_test, y_pred)
print(f'Classification Report:\n', report)
```

```python
feature_importance = pd.DataFrame({'Feature': x.columns, 'Importance': model.feature_importances_})

feature_importance = feature_importance.sort_values(by = 'Importance', ascending = False)

print('Feature Importance:\n', feature_importance)
```

```python
subset_df = df[['EPA', 'ExpPts', 'yrdline100', 'GoalToGo', 'Touchdown']]

sns.pairplot(subset_df, hue = 'Touchdown', markers = ['o', 's'], palette = {0: 'blue', 1: 'orange'})
plt.show()
```

```python
corr_matrix = subset_df.corr()
sns.heatmap(corr_matrix, annot = True, cmap = 'coolwarm', fmt = '.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (12, 8))

sns.histplot(subset_df, x = 'EPA', hue = 'Touchdown', kde = True, ax = axes[0,0], palette = {0: 'blue', 1: 'orange'})
sns.histplot(subset_df, x = 'ExpPts', hue = 'Touchdown', kde = True, ax = axes[0,1], palette = {0: 'blue', 1: 'orange'})
sns.histplot(subset_df, x = 'yrdline100', hue = 'Touchdown', kde = True, ax = axes[1,0], palette = {0: 'blue', 1: 'orange'})
sns.histplot(subset_df, x = 'GoalToGo', hue = 'Touchdown', kde = True, ax = axes[1,1], palette = {0: 'blue', 1: 'orange'})
plt.tight_layout()
plt.show()
```

```python
red_zone_plays = df[(df['yrdline100'].between(0,20)) & df['PlayType'].isin(['Run', 'Pass'])]

offensive_success_rates = red_zone_plays.groupby('PlayType')['Touchdown'].mean().sort_values(ascending = False)

plt.figure(figsize = (10,6))
sns.barplot(x = offensive_success_rates.values, y = offensive_success_rates.index)
plt.title('Touchdown Success Rates by Offensive Strategy (0-5 yards)')
plt.xlabel('Touchdown Success Rate')
plt.ylabel('Offensive Strategy')
plt.show()
```

```python
red_zone_plays = df[df['yrdline100'].between(0,5)]

offensive_success_rates = red_zone_plays.groupby('PlayType')['Touchdown'].mean().sort_values(ascending = False)

plt.figure(figsize = (10,6))
sns.barplot(x = offensive_success_rates.values, y = offensive_success_rates.index)
plt.title('Touchdown Success Rates by Offensive Strategy (0-5 yards)')
plt.xlabel('Touchdown Success Rate')
plt.ylabel('Offensive Strategy')
plt.show()
```

```python
defensive_success_rates = red_zone_plays.groupby('DefensiveTeam')['Touchdown'].mean().sort_values()

plt.figure(figsize = (10,6))
sns.barplot(x = defensive_success_rates.values, y = defensive_success_rates.index)
plt.title('Touchdown Prevention Rates by Defensive Team')
plt.xlabel('Touchdown Prevention Rate')
plt.ylabel('Defensive Team')
plt.show()
```

```python
interaction_success_rates = red_zone_plays.groupby(['PlayType', 'DefensiveTeam'])['Touchdown'].mean().unstack()

plt.figure(figsize = (12,8))
sns.heatmap(interaction_success_rates, cmap = 'viridis', annot = False, fmt = '.2f', cbar_kws = {'label': 'Touchdown Success Rate'})
plt.title('Touchdown Success Rates by Offensive Strategy and Defensive Team')
plt.xlabel('Defensive Team')
plt.ylabel('Offensive Strategy')
plt.show()
```
